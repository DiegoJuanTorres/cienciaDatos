# **Árboles de Decisión**

Los árboles de decisión son modelos predictivos que son formados mediante reglas binarias (utilizando SI/NO, <= ó >, etc) dividen los casos de un universo de datos utilizando los atributos de dichos casos como principales elementos para realizar estas divisiones.

En muchos casos un único modelo global no sirve para poder identificar una solución para poder ser aplicada a un espacio muestral. Esto se da porque las funciones "trazadas" por estos espacios muestrales son super complejos. Por ende existen alternativas en el _machine learning_ que nos permiten desagregar estos problemas y generar predictores que interacciones de una forma adecuada en base a la relación exitente entre las variables

Los árboles de decisiones pueden servir para resolver problemas tanto de regresión (variable target continua) como de clasificación (variable target discreta, generalmente dicotómica)

**Ventajas de los árboles**
- Los árboles son fáciles de interpretar.
- Los modelos que se basan en un sólo árbol son graficables, lo cual facilita la explicación de los mismo
- Al no ser un modelo paramétrico (o modelo estadístico) no debe cumplir a priori ninguna condición para que se pueda implementar.
- No son muy influenciados por los outliers.
- Se utilizan mucho a la hora de la exploración de los datos, ya que al ser fáciles de interpretar, se pueden obtener muchos _insights_ de forma rápida.

**Desventajes de los árboles**
- Una de las mayores desventajas es el bajo poder predictivo que tienen (los modelos de un sólo árbol). Esto se debe a la gran capacidad de quienen de hacer overfitting (alto sesgo) y la gran varianza que tienen.
- Tienen gran sensibilidad a las bases desbalanceadas (cuando un valor del target es sumamente superior al/los otro/s)


### _**Overfitting y Underfitting, ¿qué significan estos términos?**_
Un mal resultado en un modelo de machine learning puede ser dado por estos dos conceptos. Por eso son temas muy importantes a tener en cuenta a la hora de la construcción de un modelo predictivo. Cuando entrenamos un modelo utilizamos datos (observaciones con atributos o _features_) que a priori consideramos que son una muestra representativa del universo total con el que vamos a estar contando (tanto en la historia como en el futuro). Con tal motivo, en un modelo, creamos una función que se adapte lo mejor posible a la data histórica para que luego puedan "predecir" el comportamiento en el futuro. El overfitting (o sobreajuste) y el underfitting(o subajuste) hacen referencia al fallo del modelo al hacer "encajar" (o fit en inglés) los datos en el modelo y generalizar el conocimiento que pretendemos que adquieran.

Los modelos de Machine Learning deberán ser capaces de generalizar los conceptos. Por ejemplo, si vemos un perro labrador por primera vez en la vida y alguien nos dice: _"esto es un perro"_, luego al ver otro perro (por ejemplo un perro cuya raza es maltes) y nos preguntan: _"¿esto es un perro?"_, diremos _"No"_. Pero si en vez de mostrarnos un único perro, hubiéramos visto fotos de más de 10 razas de perros, si nos presentan otra raza no vista hasta el momento, seguro estaremos en condiciones para identificar que la nueva raza corresponde a un perro.  
Al entrenar modelos computacionales debemos utilizar un conjunto de datos de entrada que nos permita generalizar un concepto para que luego al querer hacerle la pregunta para lo que fue entrenado nos diga con mayor certeza la respuesta.

Resolver el problema del sesgo y la varianza es realmente trabajar sobre el sobreajuste y el subajuste. El sesgo se reduce y la varianza aumenta en relación a la complejidad del modelo. A medida que se agregan más y más parámetros a un modelo, la complejidad aumenta y la varianza se convierte en la principal preocupación, mientras que el sesgo disminuye constantemente.

**¿Qué es el Sesgo?**
Nos dice qué tan cerca está nuestro modelo de los datos de entrenamiento que tenemos. Al bajar el sesgo, la función encontrada en el modelo predictivo es mucho más parecido a los datos de entrenamiento, al ser más alto el sesgo, significa que no se logró una función que pueda "encajar" con los datos de entrenamiento.

**¿Y la varianza?**
Se define como la desviación de las predicciones. Nos dirá cuanto van a cambiar los resultados de las predicciones si cambian los datos puntuales que tratará de identificar. Lo que esperamos para un modelo que funciona correctamente es que si cambiamos el conjunto de datos, los valores sean predichos de la misma forma. Si el modelo tiene una alta varianza, el valor de predicción del modelo se verá afectado por los valores del conjunto de datos.
El sobreajuste ocurre cuando un modelo es sumamente complejo. Un modelo con sobreajuste tiene rendimiento muy bueno en la etapa de entrenamiento, peor al utilizar datos de validación el modelo comienza a mostrar fluctuaciones en los resultados predichos.   
El underfitting ocurre cuando un algoritmo no logra capturar la tendencia subyacente de los datos. cuando hay underfitting, el modelo no puede "modelar" de forma correcta los datos de entrenamiento y mucho menos generalizar a nuevos datos entrantes.

### **Árboles de Decisión (CART - Clasification and Regression Trees)**
En el caso de los árboles de clasificación, buscamos entrenar modelos que nos ayuden a identificar a qué categoría corresponde una observación en particular. El modelo trata de encontrar una "frontera de decisión" que le ayude a asignar una u otra categoría al dato de interés.  
La diferencia entre un árbol de decisión es que la frontera de decisión generalmente es una "recta" por ende, los datos deben ser lo más linealmente separables posibles.   
Básicamente, este algoritmo va buscando cortes en su variables para generar estas regiones para que luego estas regiones puedan ser identificadas con alguna categoría en particular o un valor continuo (en los casos de los árboles de regresión). Los árboles van eligiendo las variables y cortando de "alguna forma" para que los grupos o regiones sean lo más homogéneos posibles.

**_Índice de GINI_**
Una de las medidas para calcular la homogeneidad es el índice de Gini. Mide el grado de "impureza" de un nodo. Si el valor es 0, nos indica que el nodo es puro (o que todas las observaciones pertenecen a la misma categoría), mientras que indices mayores a 0 hasta 1 indican que el nodo posee impurezas (observaciones con distintas categorías)

$$GINI(t)= 1-{\sum_{j=1}^n P(j|t)^2}$$
Donde:
  - P(j|t) son las probabilidades de las clases
Y el GINI de la división de un nodo es:
 $$GINI_{SPLIT}= \sum_{i=1}^n \frac{N_i}{N} GINI(i)$$  

Donde:
   - Ni son los casos que quedan en el nodo i
   - N es el total de casos del nodo padre
   - GINI(i) es el índice de GINI del nuevo nodo

![join.JPG](https://www.codificandobits.com/img/posts/2021-04-11/indice-gini-ecuacion-y-ejemplos.png)

**_Entropía - Ganancia de Información_**: La entropía, mide la incertidumbre que posee una fuente de información. Otra forma de ver la entropía es como la cantidad de información promedio que poseen los símbolos utilizados
$$Entropía(t) = - \sum_{j=0}^n p(j|t) log_2  p(j|t)$$
Donde:
  - P(j|t) son las probabilidades de las clases.  

Para el cálculo de la ganancia de la información debemos realizar lo siguiente:
$$GAIN_{SPLIT} = Entropía(p) - (\sum_{i=1}^k \frac{n_i}{n} Entropía(i))$$
Donde: 
- k es la cantidad de nodos que se desprenden del nodo padre.
- ni es la cantidad de observaciones en cada uno de los nodos hijos
- n es la cantidad total de observaciones del nodo padre
- Entropía(i) es la entropía del nodo hijo i  

Los árboles elijen el mejor según maximice la GANANCIA.
Para evitar peque nas particiones se puede utilizar una penalización a la
hora de utilzar la ganancia de información. Esta penalización se la conoce
como Gain Ratio o Razón de la ganancia y se calcula de la siguiente forma:  
$$Gain_{ratio} = \frac{GAIN_{split}}{Split_{info}}$$
Donde:  
- $Split_{info} = -\sum_{i=1}^k \frac{n_i}{n}log\frac{n_i}{n}$
- k es la cantidad de nodos que se desprenden del nodo padre.
- ni es la cantidad de observaciones en cada uno de los nodos hijos
- n es la cantidad total de observaciones del nodo padre




### _**Árboles de Clasificación**_
Para la construcción de este tipo de modelo se utiliza el mismo método de _recursive binary splitting_ visto anteriormente. El único cambio que se realiza es la función que vamos a estar optimizando. Existen varias alternativas de funciones objetivo, estas funciones buscarán tener los nodos más puros / homogéneos posibles.  
Las funciones más utilizadas son:

**_Classification Error Rate_:** es la proporción de observaciones que no pertenecen a la clase más común del nodo.

**Función de costo para la elección de la mejor partición**
Para saber cuál es la mejor partición se define una función de costo que asigna un puntaje al nodo padre, usando los promedios ponderados de los índices individuales de sus nodos hijos.

La función de costo será cada una de las métricas multiplicando el % de casos que cayó en cada uno de los nodos. De esa forma se pondera cada una de las métricas de impureza. Luego se suman las métricas de impureza para tener la función de costo total.

Independientemente de la medida que se utilice para buscar el punto óptimo. El algoritmo funciona de la siguiente forma:
1) Para cada posible división se calcula el valor de la medida para los nodos resultantes
2) Se suman los dos valores, ponderando cada nodo por la fracción de observaciones que posee.
3) La división con menor o mayor valor (dependiendo la métrica empleada) es seleccionada como punto de corte óptimo.

Las métricas más utilizadas son Gini y Entropía, ya que son las que poseen mayor sensibilidad a la homogeneidad de los nodos.

### _**Árboles de regresión**_
Este tipo de modelos son aquellos que se aplican cuando la variable a predecir (variable target) es continua. Como funciona un árbol, las diferentes observaciones de una base (también conocidas como registros) van bifurcándose en el árbol mediante valores de sus atributos, éstas llegan a un nodo terminal. La predicción del árbol será la media de los valores de las observaciones que están ubicadas en cada uno de los nodos, por tal motivo, cuando un nuevo valor entre al modelo, pasará por cada una de las ramas de pendiendo de cada valor de sus atributos y se alojará en un nodo terminal, tomando el valor predicho como la media del nodo.

#### ¿Cómo se entrena un árbol?
De una forma sencilla, los árboles (tanto de regresión como de clasificación) se entrenan en dos etapas:
1) Se realiza una división sucesiva del espacio de los predictores generando regiones no solapantes (mutuamente excluyentes) denominados nodos terminales. Esto quiere decir que una observación no puede estar en dos nodos al mismo tiempo.
2) Predicción de la variable respuesta en cada una de las regiones (nodos terminales)

Estas dos etapas parecen sencillas, pero se debe crear una metodología para poder crear estas regiones. Es decir, mediante una metodología, se deben elegir qué predictores son los mejores (atributos de cada observación) y cuál es el punto de corte óptimo (valor que mejora dicho corte). Aquí comienza la diferenciación entre los árboles de regresión y los de clasificación.  
En los árboles de regresión, el método más utilizado para identificar estas divisiones es el _Residual Sum of Squares_(o suma de residuos al cuadrado, el residuo es la diferencia entre el valor real y el predicho). El objetivo es encontrar las _J_ regiones que minimicen esta suma:

$$RSS = \sum_{i=1}^{J}\sum_{i∈R_j} (y_i - \hat{y_{R_i}})^2$$

Donde $y_{R_i}$ es la media de la región $R_j$  
Para decirlo en otras palabras, se trata de buscar una distribución de regiones en donde la sumatoria de las desviaciones al cuadrado entre las observaciones y la media de la región a la que pertenece sea el valor menor posible. Como no se pueden considerar todas las posibles particiones se utiliza un método denominado _Recursive Binary Splitting_ o división binaria recursiva.

#### _Recursive Binary Splitting_
El objetivo del método es encontrar en cadad iteración el predictor $x_j$ y el punto de corte $s$ (o umbral) tal que, si se distribuyen las observaciones en las regiones {X|Xj<s} y {X|Xj≥s}, se consigue la mayor reducción posible en el RSS. 

El algoritmo posee los siguientes pasos:
1) El proceso se inicia en lo más alto del árbol, donde todas las observaciones pertenecen a la misma región
2) Se identifican todos los posibles puntos de corte (umbrales) s para cada uno de los predictores (X1, X2,…, Xp). En el caso de predictores cualitativos, los posibles puntos de corte son cada uno de sus niveles. Para predictores continuos, se ordenan de menor a mayor sus valores, el punto intermedio entre cada par de valores se emplea como punto de corte.
3) Se calcula el RSS total que se consigue con cada posible división identificada en el paso 2.
$$RSS = RSS_{split 1}+RSS_{split 2}$$
4) Se selecciona el predictor Xj y el punto de corte S que resulta en el menor RSS total, es decir, que da lugar a las divisiones más homogéneas posibles. Si existen dos o más divisiones que consiguen la misma mejora, la elección entre ellas es aleatoria.
5) Se repiten de forma iterativa los pasos 1 a 4 para cada una de las regiones que se han creado en la iteración anterior hasta que se alcanza alguna norma de stop. Algunas de las más empleadas son: que ninguna región contenga un mínimo de n observaciones, que el árbol tenga un máximo de nodos terminales o que la incorporación del nodo reduzca el error en al menos un % mínimo.

Está metodología conlleva dos hechos:

- Que cada división óptima se identifica acorde al impacto que tiene en ese momento. No se tiene en cuenta si es la división que dará lugar a mejores árboles en futuras divisiones.

- En cada división se evalúa un único predictor haciendo preguntas binarias (si, no), lo que genera dos nuevas ramas del árbol por división. A pesar de que es posible evaluar divisiones más complejas, hacer una pregunta sobre múltiples variables a la vez es equivalente a hacer múltiples preguntas sobre variables individuales.

_Nota: Los algoritmos que implementan recursive binary splitting suelen incorporar estrategias para evitar evaluar todos los posibles puntos de corte. Por ejemplo, para predictores continuos, primero se crea un histograma que agrupa los valores y luego se evalúan los puntos de corte de cada región del histograma (bin)._

# **Ensembles**
Los modelos estadísticos y de Machine Learning poseen un problema que es el equilibrio entre el sesgo (bias) y la varianza (variance)  
El concepto de **sesgo** hace referencia a cuánto en promedio los valores con respecto a los valores reales. Es decir, nos dice qué tan real es la relación que encontró entre los predictores y la variable respuesta.  
El término varianza nos dice cuánto cambia el modelo dependiendo de los datos utilizados en el entrenamiento. Lo que uno espera que un modelo pueda _**generalizar**_ su comportamiento, es decir, que con observaciones sucesivas (diferentes a las que se utilizó en el entrenamiento) el modelo no varíe demasiado sus resultados.  
A medida que se aumenta la complejidad del modelo, éste logra adaptarse mejor a la realidad que se debe modelar, no obstante, si esta complejidad no es medida, puede aparecer un concepto denominado _overfitting_, esto significa que el modelo se ajusta tanto a los datos de entrenamiento que no es capaz de identificar nuevas observaciones. Por tal motivo, el mejor modelo será aquel que pueda equilibrar de la mejor manera el sesgo y la varianza.  
La gran pregunta será, ¿cómo puedo controlar el sesgo y la varianza? Los árboles pequeños tienen poca varianza pero consiguen representar bien la relación entre las variables (alto sesgo). Por el contrario, árboles grandes se ajustan mucho a los datos de entrenamiento, por ende el sesgo es bajo pero la varianza es muy alta. Un método para resolver esta problemática es trabajar con métodos de _ensembles_.  
Los **ensembles** son métodos que combinan múltiples modelos en uno sólo que tienen como objetivo lograr un equilibrio entre el sesgo y la varianza. Los dos ensembles más utilizados son:
-  **Bagging**: Se generan muchos modelos, con un subconjunto distinto de datos para cada uno de los mismo. (Estos subconjuntos son armados en base a los datos de entrenamiento). Todos los modelos aportan a la hora de dar el _veredicto_ final, siendo así el valor final como la media de todas las predicciones(variables continuas) o la clase más frecuente (variables categóricas). El modelo más utilizado de esta técnica se denomina **Random Forest**
- Boosting: Se ajustan secuencialmente modelos sencillos, denominados _weak learners_. Cómo valor final se toma la media cuando la variable a predecir es una variable continua o la clase más votada (variable categórica).  
Ambos atacan el mismo problema, la optimización entre el sesgo y la varianza. No obstante cada modelo trabaja de una forma diferente.
- En el bagging se emplean modelos con muy poco sesgo pero muchísima varianza, lo que agregando dichos modelos se loggra reducir dicha varianza sin apenas modificar el sesgo.
- En cambio en el boosting se emplean modelos con muy poca varianza pero mucho sesgo, que ajustanto secuencialmente, los modelos reducen.

En el bagging la forma de tomar muestras aleatorias sobre la base de entrenamiento se realiza sobre la técnica de bootstraping. En los modelos boosting se ajustan los pesos de cad modelo en base a cada iteración.

Estos tipos de modelos tienen un gran resultado en comparación de otros modelos más clásicos (por ejemplo regresiones logísticas, etc) pero se pierde un poco la interpretabilidad de los modelos. 

## **Bagging** (Bootstrap aggregation)
Dadas $n$ mestras de observaciones independientes $Z_1, .., >_n$, cada una con una varianza $\epsilon^2$, la varianza de la media de las observaciones es de $\epsilon^2/n$. Dicho de otra forma, al promediar un con junto de observaciones, se reduce la varianza.  
El Random Forest es una implementación del método Bagging con una diferencia, utiliza además una selección m predictores para cada una de las muestras armadas.


## **Boosting**
Cómo se dijo anteriormente, la idea de este método es ajustar secuencialmente modelos débiles. De esta forma, cada nuevo modelo utiliza la historia del modelo para poder corregir sus propios errores. A diferencia del bagging, no se hacen muestreos aleatorios.  
Todas las implementaciones de este método se caracterizan por tener muchos hiperparámetros asociados, los más importantes son:
- Número de _weak learners_ (cuántas iteraciones), a diferencia del bagging, en el boosting puede generar overfitting.
- Learning Rate: controla la influencia que tiene cada uno de los _weak learners_ en el conjunto total. Dicho de otras palabras, marca el ritmo en el que aprende el modelo. Suele usarse un valor que va entre el 0.001 y el 0.01. Si el valor es más chico, se necesitarán más iteraciones, pero menor va a ser el riesgo de overfitting. En contraposición el modelo tardará mucho más en aprender.
- El valor máximo del árbol permitido.

### **Extreme Gradient Boosting**
XGBoost (por sus siglas en eXtreme Gradient Boosting) es uno de los algoritmos de ensembles más elegidos por los equipos que desarrollan soluciones
para las competencias propuestas en el Sitio Kaggle (sitio más importante en competencias de Machine Learning). Este ensemble proviene de la familia del meta algoritmo Boosting. El XGBoost es un modelo boosting que además de utilizar dicha técnica optimiza la función de pérdida utilizando el descenso del gradiente. Para poder entender el algoritmos es necesario poder explicar qué es el Descenso del Gradiente. Todos los algoritmos Gradient
boosting deben poseer tres elementos:
- Una función de pérdida a optimizar.
- Un algoritmo base débil (por ejemplo árboles de decisión)
- Un modelo aditivo para poder sumar cada uno de los estimadores o algoritmos base.
Los árboles se agregan de a uno a la vez, y los árboles que ya fueron agregados no cambian. Para poder definir los parámetros que los árboles tendrán se utiliza 
el concepto de descenso de gradiente para poder minimizar la función de pérdida. El método de optimización del descenso del gradiente lo veremos a continuación.


#### Métodos de optimización
Estimación de Máxima Verosimilitud: Como se mencionó anteriormente la idea del método es conseguir el mejor valor de estimación sobre un parámetro para poder maximizar la probabilidad de obtener una muestra determinada. Sabiendo que la probabilidad de un suceso es P(X) = p × (1 − p).  
Para poder explicar el método vamos a usar un ejemplo: Lanzaremos una moneda 5 veces y veremos su secuencia: Seca, Cara, Cara, Cara, Seca.
P(Seca, Cara, Cara, Cara, Seca) = $(1 − p) * p * p * p * (1 − p) = p^3 *(1 − p)^2$
Con esta secuencia, el p de Cara es del 0.6 $\frac{3}{5}$ y por ende la probabilidad es: P(Seca, Cara, Cara, Cara, Seca) = $0,6^3(1-06)^2 = 0,0346$ Para esta muestra lo que vamos a calcular son todos p posibles y su probabilidad.

Haciendo esta comparación podemos observar que para esta muestra de
observaciones, el valor donde se encuentra la mayor probabilidad es justamente cuando p=0.6. Entonces el parámetro pˆ será 0.6. Cuando vemos este ejemplo podemos llegar a entender que el parámetro estimado por máxima verosimilitud es igual a la frecuencia relativa del evento que deseamos estimar. Ahora lo que trataremos de hacer es generalizar la fórmula y realmente entender si esta experimentación es correcta.  
Prob(k caras en n lanzamientos)$ = p^k (1-p)^{n-k} = L(p)$

## Parámetros Boosting
El boosting es un ensamble de *learners* débiles (generalmente árboles de decisión) que trabajan en conjunto para poder lograr resultados mucho más robustos en su conjunto. 
La forma en que se generan es secuencial, es decir, se genera un árbol de, el mismo es evaluado, luego se generan *pesos* en cada una de las observaciones, las observaciones que fueron érroneas en sus predicciones, se les dará mayor peso, las que fueron acertadas, menor. Esto se hace para que el árbo siguiente haga más hincapié en aquellas observaciones que fueron mal predichas.

Existen varias implementaciones de este algoritmo:

* XGBoost (Extreme Gradient Boosting): Es uno de los algoritmos que estuvo dominando las competencias de data mining a nivel global. Creado por Tianqui Chen y Carlos Guestrin, logra muy buenos resultados y excelentes generalizaciones. En el siguiente [link](https://xgboost.readthedocs.io/en/latest/) podrán encontrar toda su documentación.
* LightGBM: Es una implementación de un modelo boosting creado por Microsoft, mejora el XGBoost generando muy buenos resultados. Para ver la documentación de dicha implementación realizar click en el siguiente [enlace](https://lightgbm.readthedocs.io/en/latest/pythonapi/lightgbm.LGBMClassifier.html#lightgbm.LGBMClassifier)

Los parámetros para ajustar del **XGBoost** son:

* base_score: Este parámetro es el score que se le asigna inicialmente a todas las predicciones.
* Booster: Este parámetro nos indica que se desea usar como *learner* base. Las tres opciones de este algoritmo son: gbtree, dart (ambos se basan en árboles) y gblinear (funciones lineales)
* colsample_bylevel: Se realiza un muestreo de las columnas a elegir encada uno de los niveles del árbol. Es decir, cada vez que crece un nivel el árbol, se elije una característica de una submuestra.
* colsample_bynode: Por cada nodo del árbol se realiza una muestra de las posibles variables para elegir y realizar un nuevo corte.
* colsample_bytree: Por cada árbol nuevo, se realiza una selección de variables.
* gamma: Es igual al min_split_loss. Es la mínima reducción de pérdidas que se requiere para poder realizar un nuevo split en un nodo hoja. Cuanto más alto sea el valor de gamma, más conservador será el árbol. 
* importance_type: Es el tipo de feature importance. Los valores son gain, weight o total_gain.
* learning_rate: Es la tasa de aprendizaje. Si el valor es muy chico, los pasos para buscar los mínimos de las funciones serán más cortos, si el valor es muy grande, los pasos serán más largos.
* max_delta_step: Es el máximo delta que se puede tener cuando se estimen los pesos de los árboles.
* max_depth: Profundidad máxima de los árboles bases.
* missing: El valor con el que se completarán los valores faltantes.
* n_estimators: Cantidad de *learners* base que tendrá el modelo final.
* n_jobs: Número de jobs (threads) que correrán en paralelo.
* random_state: Valor de la semilla para que puedan ser reproducidos los resultados.
* reg_alpha: Regularización L1.
* reg_lambda: Regularización L2.
* scale_pos_weight: Balanceo entre los pesos de las clases positivas o negativas.
* subsample: Es un ratio de muestreo para las instancias de entrenamiento.
* verbosity: Valor por default=1. Este parámetro nos indicará que mensajes se mostrarán por pantalla. El valor 0 indica falta de mensajes, 1 solo advertencias, 2 información, 3 información de debug.

## Creo el gráfico _Feature Importance_

Es una técnica que nos ayuda a calcular un score para todos los _features_ (o características). Simplemente representa la importancia que posee cada característica sobre el modelo. A mayor score significa que posee mayor efecto sobre el modelo.


```python
#La importancia de las variables o feature importance nos permite identificar entre todos los modelos débiles que se generan las variables que más aparecen y te los pondera para 
#dar una explicación de las variables
importances = xgbModel.feature_importances_
importances = importances / importances.sum()
indices = np.argsort(importances)[::-1][0:20]
names = [X_train.columns[i] for i in indices]
plt.figure()
plt.title("Feature Importance model 1")
plt.bar(range(len(names)), importances[indices])
plt.xticks(range(len(names)), names, rotation=90)
```
